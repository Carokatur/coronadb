# Pre-Process data "ärztliches / nicht-ärztliches Personal"

```python
import pandas as pd
import numpy as np
```

```python
# read raw data from csv
df = pd.read_csv('C:/Users/mriedesel/Desktop/coronadb/aerztliches-nicht-aerztliches-personal-nach-bundesland.csv', sep=';', decimal=',')
```

```python
df.head()
```

```python
# create relative values per Bundesland

df['Ärztlicher Dienst_20171231_Rel'] = df['Ärztlicher Dienst_20171231'] / df['Insgesamt_20171231']
df['Nicht-ärztlicher Dienst_20171231_Rel'] = df['Nicht-ärztlicher Dienst_20171231'] / df['Insgesamt_20171231']
df['Auszubildende_20171231_Rel'] = df['Auszubildende_20171231'] / df['Insgesamt_20171231']
df['Ärztlicher Dienst_2017_Avg_Rel'] = df['Ärztlicher Dienst_2017_Avg'] / df['Insgesamt_2017_Avg']
df['Nicht-ärztlicher Dienst_2017_Avg_Rel'] = df['Nicht-ärztlicher Dienst_2017_Avg'] / df['Insgesamt_2017_Avg']
```

```python
df.head()
```

```python
# Write to csv
df.to_csv('C:/Users/mriedesel/Desktop/coronadb/aerztliches-nicht-aerztliches-personal-nach-bundesland_edited.csv', index=False)
```
